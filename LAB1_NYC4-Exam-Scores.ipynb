```python
#https://github.com/IST256/learn-python/blob/master/content/lessons/12/Now-You-Code/NYC4-Exam-Scores.ipynb
import pandas as pd
URL='https://raw.githubusercontent.com/mafudge/datasets/master/exam-scores/exam-scores.csv'
```

```python
df = pd.read_csv(URL)
df.head(5)aaaa
```

```python
Class_Section = 'M01'
Exam_version = 'A'
```

```python
#With valid inputs it should filter the DataFrame to only those rows matching the input (for example section M01, version A)

df_filter = df[(df['Class Section'] == Class_Section) & (df['Exam Version'] == Exam_version) ]
df_filter
```

```python
#And display only the Student Score and Letter Grade for each student matching the inputs.
df_few_fields = df_filter[['Student Score','Letter Grade']]
df_few_fields
```

```python
Class_Section = input("Enter a section ['M01', 'M02'] : ")
Exam_version = input("Enter an exam version ['A', 'B', 'C', 'D'] :" )

df_filter = df[(df['Class Section'] == Class_Section) & (df['Exam Version'] == Exam_version) ]
df_few_fields = df_filter[['Student Score','Letter Grade']]
print("For Student Score {} and Letter Grade {} ".format(Class_Section, Exam_version))
df_few_fields
```
