```python
import pandas as pd
import requests as req
import warnings
KEY='7a152016ec0da61f6cf528d5ace18d1a'
LOCATIONS='43.0481221,-76.1474244'
URL="https://api.darksky.net/forecast/{}/{}".format(KEY,LOCATIONS)
```

```python
resp = req.get(URL)
dict = resp.json()
#print(dict['daily'])
df_sarkSky=pd.DataFrame(dict['daily']['data'])
df_sarkSky.head(5)
```

```python
warnings.simplefilter("ignore")
df_selected = df_sarkSky[['time','temperatureHigh','temperatureLow']]
df_selected['time'] = pd.to_datetime(df_selected['time'],unit='s')
df_selected
```
