```python
import pandas as pd

sh = pd.read_csv('https://raw.githubusercontent.com/mafudge/datasets/master/superhero/superhero-movie-dataset-1978-2012.csv')
sh.head()
```

```python
sh.columns = [ 'year', 'title', 'comic', 'imdb', 'rt', 'composite', 'opening_weeked_bo', 'avg_ticket_price', 'opening_weekend_attend', 'us_pop_that_year']
sh.head()
```

```python
## let's see that as a percentage of the total
## what are the ratios in the last 10 years of data ?
print(sh['comic'].value_counts())
sh2 = sh.dropna()
print(sh2['comic'].value_counts())
print(sh[sh['year']>2002]['comic'].value_counts())
```

```python
# feature engineering
sh2['pct_of_pop'] = sh2['opening_weekend_attend'] /sh2['us_pop_that_year']
sh2.head()
```

```python
import pandas as pd
url = 'https://ischool.syr.edu/classes/?page=1'
df1 = pd.read_html(url)
df1[0]
```

```python
url = 'https://ischool.syr.edu/classes/?page='
classes = pd.DataFrame()
for i in range(1,6):
    df1 = pd.read_html(url + str(i))
    classes = classes.append(df1[0],ignore_index=True)

classes
```

```python
classes['subject']=classes['Course'].str[0:3]
classes['number']=classes['Course'].str[3:]
classes
```

```python
classes['type']=''
classes['type'][classes['number'] < '500'] = 'Ugrade'
classes['type'][classes['number'] >= '500'] = 'Grade'

classes
```

```python
classes['type'].value_counts()
df_friday = classes[(classes['Day'].str.find('F')>=0) & (classes['type'] == 'Ugrade')]
#df_friday = classes[classes['Day'] = 'F']
df_friday['Day'].value_counts()
```

```python
df_friday_noLabs = df_friday[~df_friday['Title'].str.startswith('LAB:')]
#df_friday_noLabs = df_friday[df_friday['Title'].str.startswith('LAB:')]

df_friday_noLabs.count()
```
